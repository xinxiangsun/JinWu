```python

from jinwu.core.time import Time

print(Time('2015-06-30 23:59:59', format='iso', scale='utc').swift)
print(Time('2015-06-30 23:59:60', format='iso', scale='utc').swift)
print(Time('2015-07-01 00:00:00', format='iso', scale='utc').swift)
```

```python
from jinwu.core.time import Time
import astropy.units as u

baseline = Time('2015-07-01 00:00:00', format='iso', scale='utc')
for i in range(5):
    t = baseline + i * u.s
    print(i, (t.swift - baseline.swift))
```
